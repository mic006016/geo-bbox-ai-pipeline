```python
from ultralytics import YOLO

# 1. 모델 로드
model = YOLO('yolov8s.pt')

# 2. 전이 학습 시작
results = model.train(
    data='/content/yolo_dataset2/dataset2.yaml',
    epochs=50,
    imgsz=512,
    batch=64,
    device=0,
    workers=8,
    amp=True,
    project='geo_ai',
    name='road_damage_v1',
    save=True,
    patience=10
)
```
